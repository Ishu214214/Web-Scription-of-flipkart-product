```python
from bs4 import  BeautifulSoup as soup
from urllib.request import urlopen as uopen
my_url="https://www.flipkart.com/search?q=iphon&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

x=uopen(my_url)
html_page=x.read()
x.close()
new_page=soup( html_page , "html.parser" )

```

```python
new_page
```

```python
cout = new_page.findAll( "div" ,{'class' : "_13oc-S"} )
cout
```

```python
len(cout)
```

```python
soup.prettify(cout[0])
```

```python
count1 =cout[0]
count1
```

```python
title=count1.div.img["alt"]
print(title)
```

```python
ratings = count1.findAll('span'  , { 'class' : '_2_R_DZ'})
print(ratings[0].text)
```
